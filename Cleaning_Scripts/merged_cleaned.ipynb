```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

```python
# File to Load
node_path = Path("Datasets/NODE.csv")
accident_path = Path("Datasets/ACCIDENT.csv")
```

```python
# Read the CSV file
node_data = pd.read_csv(node_path, low_memory=False)
accident_data = pd.read_csv(accident_path, low_memory=False)
```

```python
# Store it in a Dataframe
node_data_df = pd.DataFrame(node_data)
node_data_df
```

```python
# Store it in a Dataframe
accident_data_df = pd.DataFrame(accident_data)
accident_data_df.head()
```

```python
left_join = pd.merge(node_data_df, accident_data_df, on='ACCIDENT_NO', how='left')
left_join
```

```python
# Find columns with missing values
columns_with_missing_values = left_join.isnull().any()

# Display the columns with missing values
columns_with_missing_values
```

```python
# Drop columns with missing values
df_dropped = left_join.drop(columns=columns_with_missing_values[columns_with_missing_values].index, inplace=True)
df_dropped
```

```python
# Storing file in Data folder
merged_data_output_path = Path("Cleaned_Datasets/merged_cleaned.csv")
left_join.to_csv(merged_data_output_path, index=None, header=True)
```
